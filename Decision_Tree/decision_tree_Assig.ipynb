```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('loan_approval_dataset.csv')
df
```

```python
x=df.drop(columns='Loan_Approved')
y=df.Loan_Approved
```

```python
obj_cols=x.select_dtypes(include='object').columns
obj_cols
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
xtrain[obj_cols].nunique()
```

```python
encode=OneHotEncoder(sparse_output=False, handle_unknown='ignore')
values=encode.fit_transform(xtrain[obj_cols])
cols=encode.get_feature_names_out()
```

```python
xtrain[cols]=values
```

```python
xtrain.drop(columns=obj_cols,inplace=True)
xtrain
```

```python
v=encode.transform(xtest[obj_cols])
c=encode.get_feature_names_out()
xtest[c]=v
xtest.drop(columns=obj_cols,inplace=True)
xtest
```

```python
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
ypred_train=model.predict(xtrain)
```

```python
print(classification_report(ytrain,ypred_train))
```

```python
ypred_test=model.predict(xtest)
```

```python
print(classification_report(ytest,ypred_test))
```

- here when i evaluated my model with ytrain, the accuracy score shows 100% , so it is underfit , the model has memorized the train data
- in model evaluation using the ytest, the accuracy scores dropped to 52% which shows the model performs better only on the seen data not on the unseen data so it is not generalized well
