```python
import pandas as pd
```

```python
df=pd.read_csv('energy_data.csv')
df
```

```python
df.columns
```

```python
df.corr()
```

```python
df=df.drop(columns=['Average Temperature'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
```

```python
x=df[['Square Footage', 'Number of Occupants', 'Appliances Used']]
y=df['Energy Consumption']
```

```python
model=LinearRegression()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
ypred_train=model.predict(X_train)
score=r2_score(Y_train,ypred_train)
score
```

```python
mae=mean_absolute_error(Y_train,ypred_train)
mae
```

```python
rmse=root_mean_squared_error(Y_train,ypred_train)
rmse
```

```python
ypred_test=model.predict(X_test)
score1=r2_score(Y_test,ypred_test)
score1
```

```python
mae1=mean_absolute_error(Y_test,ypred_test)
mae1
```

```python
rmse1=root_mean_squared_error(Y_test,ypred_test)
rmse1
```

- As the values of MAE and RMSE are far so the model has failed to capture the outliers
