```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

- model_selection is module and train_test_split is a method

```python
import pandas as pd
df=pd.read_csv('Salary_Data.csv')
```

```python
x=df[['YearsExperience']]    
y=df['Salary']
```

## Syntax of train_test_split
- Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,train_size=0.8, random_state=42)
- train_size is the value which we tell to take 80% of data for training and 20% for test data.
- random_state is given to make the values constant for the column if we dont use the random_state it will change the values for each run.

- We use the variables to store the 4 values which are returned by the train_test_split the splitted data using unpacking method

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
Xtrain
```

```python
model=LinearRegression()
```

```python
model.fit(Xtrain,Ytrain)
```

- model.fit(Xtrain,Ytrain) learns the pattern from Xtrain and Ytrain and stores the information inside the object "model".
- model.fit(Xtrain,Ytest) ERROR -> Solution (check the samples) model learns the pattern from same no. of samples valid (80% Samples, 80% Samples)

```python
Ypred=model.predict(Xtrain)
```

- Here in this step we will evaluate the performance of model to check how the model is predicting before giving the new input.

- metrics is a module which contains the model evaluation methods only for regression not for only Linear regression and it is for all regression algorithms
- Error is the difference the Ytrain and Ypred(model.predict(Xtrain))
- Error is the mistake done by the model.
