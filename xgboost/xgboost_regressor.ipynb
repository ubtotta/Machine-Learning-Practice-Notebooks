```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
df=pd.read_csv('employee_performance.csv')
df.drop(columns='Employee_ID')
```

```python
x=df.drop(columns='Salary')
y=df.Salary
```

```python
obj_cols=x.select_dtypes(include='object').columns
obj_cols
```

```python
x[obj_cols]=x[obj_cols].astype('category')
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
model=XGBRegressor(enable_categorical=True,n_estimators=200,random_state=42)
model.fit(xtrain,ytrain)
```

```python
ypred_train=model.predict(xtrain)
```

```python
print(r2_score(ytrain,ypred_train))
print(root_mean_squared_error(ytrain,ypred_train))
print(mean_absolute_error(ytrain,ypred_train))
```

```python
ypred_test=model.predict(xtest)
```

```python
print(r2_score(ytest,ypred_test))
print(root_mean_squared_error(ytest,ypred_test))
print(mean_absolute_error(ytest,ypred_test))
```
